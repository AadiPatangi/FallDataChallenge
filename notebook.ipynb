```python

# Assuming `incomes` is your original DataFrame
# Filter the columns containing 'Total!!Estimate' as you did
income_data = incomes.filter(regex="Total!!Estimate")[12:13]

# Clean the column names by removing the 'ZCTA5' and '!!Total!!Estimate' part
income_data.columns = income_data.columns.str.extract(r'(\d{5})')[0]

# 1. Extract the ZIP codes from the column names
zip_codes = income_data.columns

# 2. Create a dictionary to store the ZIP code to income mapping
zip_income_mapping = {}

# 3. Iterate through each cleaned column name (which is now a ZIP code) and get the corresponding income value
for zip_code in zip_codes:
    # Access the income value from the corresponding column and row (0th index for the single row in the DataFrame)
    income_value = income_data[zip_code].iloc[0]
    
    # Add the ZIP code and income value to the dictionary
    zip_income_mapping[zip_code] = income_value

# Now, zip_income_mapping contains the ZIP code and income value pairs
print(zip_income_mapping)
```

```python
# Extract the counts of ZIP codes from data_cleaned['ZIPCode']
zip_counts = data_cleaned['ZIPCode'].value_counts()
# Clean the column names in income_data by stripping the 'ZCTA5' prefix
cleaned_columns = [col.split(' ')[-1] for col in income_data.columns]  # Extract the ZIP code
income_data.columns = cleaned_columns  # Update the column names of income_data

# Clean the zip_codes list to match the same format (ensure they are strings)
zip_codes = [str(zip_code).strip() for zip_code in zip_codes]
print(income_data.columns)  # Check the column names in the income data


# Create a new dictionary to combine income and count for each ZIP code


zip_info_mapping = {}

for zip_code in zip_codes:
    try:
        # Ensure the ZIP code exists as a column in the income_data DataFrame
        if str(zip_code) in income_data.columns:
            income_value = income_data[str(zip_code)].iloc[0]  # Get the income value for the ZIP code
            count_value = data_cleaned['ZIPCode'].value_counts().get(zip_code, 0)  # Get count of records for the ZIP code
            
            # Map the ZIP code to its income and count values
            zip_info_mapping[zip_code] = {'income': income_value, 'count': count_value}
        else:
            print(f"Warning: Missing data for ZIP code {zip_code}")
    except Exception as e:
        print(f"Error processing ZIP code {zip_code}: {e}")


```

```python
# Step 1: Normalize ZIP codes in data_cleaned to the first 5 digits
data_cleaned['ZIPCode'] = data_cleaned['ZIPCode'].astype(str).str[:5]  # Take first 5 digits

# Step 2: Clean the income values in zip_info_mapping (ensure they are numeric)
for zip_code, data in zip_info_mapping.items():
    if isinstance(data['income'], str):
        income_cleaned = data['income'].replace(',', '')  # Remove commas
        try:
            zip_info_mapping[zip_code]['income'] = float(income_cleaned)  # Convert income to float
        except ValueError:
            zip_info_mapping[zip_code]['income'] = None  # If there's an issue, set income as None
    elif isinstance(data['income'], (int, float)):
        zip_info_mapping[zip_code]['income'] = float(data['income'])  # Ensure it's a float

# Step 3: Normalize the keys in zip_info_mapping to 5 digits
zip_info_mapping = {str(zip_code).strip()[:5]: value for zip_code, value in zip_info_mapping.items()}

# Step 4: Get all ZIP codes in data_cleaned
unique_zip_codes_in_data = data_cleaned['ZIPCode'].unique()
print("Unique ZIP codes in data_cleaned:", unique_zip_codes_in_data)

# Step 5: Log missing ZIP codes (those that are in data_cleaned but not in zip_info_mapping)
missing_zip_codes = [zip_code for zip_code in unique_zip_codes_in_data if zip_code not in zip_info_mapping]
print("Missing ZIP codes (not in zip_info_mapping):", missing_zip_codes)

# Step 6: Add missing ZIP codes to zip_info_mapping with a count of 0 and income as None
for zip_code in missing_zip_codes:
    zip_info_mapping[zip_code] = {'income': None, 'count': 0}  # Add missing zip codes with income None and count 0

# Step 7: Count occurrences of ZIP codes in data_cleaned
zip_code_counts = data_cleaned['ZIPCode'].value_counts()

# Step 8: Update counts in zip_info_mapping for all ZIP codes
for zip_code in zip_info_mapping:
    if zip_code in zip_code_counts:
        zip_info_mapping[zip_code]['count'] = zip_code_counts[zip_code]
    else:
        zip_info_mapping[zip_code]['count'] = 0  # If no match, ensure count is 0

# Step 9: Verify the updated zip_info_mapping
print("Updated Zip Info Mapping:", zip_info_mapping)
```

```python
import matplotlib.pyplot as plt

# Prepare data for plotting: Extract income values and counts
income_values = [zip_info_mapping[zip_code]['income'] for zip_code in zip_info_mapping]
count_values = [zip_info_mapping[zip_code]['count'] for zip_code in zip_info_mapping]

# Create the plot
plt.figure(figsize=(10, 6))
plt.scatter(income_values, count_values, color='b', alpha=0.5)

# Labeling the axes and the title
plt.xlabel('Income ($)', fontsize=12)
plt.ylabel('Counts', fontsize=12)
plt.title('Income vs Counts by ZIP Code', fontsize=14)

# Display the plot
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats

# Prepare data for plotting: Extract income values and counts
income_values = [zip_info_mapping[zip_code]['income'] for zip_code in zip_info_mapping]
count_values = [zip_info_mapping[zip_code]['count'] for zip_code in zip_info_mapping]

# Remove any entries where income or count is None or NaN, and remove points with less than 10000 calls
filtered_data = [(income, count) for income, count in zip(income_values, count_values) if income is not None and count is not None and count >= 10000]

# Separate the cleaned data back into income and count lists
filtered_income_values, filtered_count_values = zip(*filtered_data)

# Create the plot
plt.figure(figsize=(10, 6))
plt.scatter(filtered_income_values, filtered_count_values, color='b', alpha=0.5)

# Fit a linear model (least squares)
slope, intercept, r_value, p_value, std_err = stats.linregress(filtered_income_values, filtered_count_values)

# Plot the regression line
plt.plot(filtered_income_values, np.array(filtered_income_values)*slope + intercept, color='r', label='Linear Fit')

# Labeling the axes and the title
plt.xlabel('Income ($)', fontsize=12)
plt.ylabel('Counts', fontsize=12)
plt.title('Income vs Counts by ZIP Code Baltimore (Filtered for Calls >= 10000)', fontsize=14)

# Set the limits for the axes
plt.xlim(40000, max(filtered_income_values)*1.1)  # Start x-axis at 40,000 and extend to the max income value
plt.ylim(0, max(filtered_count_values)*1.1)  # Start y-axis at 0 and extend to the max count value

# Display the plot
plt.grid(True)
plt.legend()
plt.show()

# Output R-squared and p-value
print(f"R-squared: {r_value**2:.4f}")
print(f"P-value: {p_value:.4f}")
```
