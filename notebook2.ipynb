```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'data' is your DataFrame containing the ACTTIME column
# If the ACTTIME column contains integer times like 1142, 915, etc.
Times = data['ACTTIME']

# Convert integer time format (HHMM) to HH:MM string format
def convert_to_time(x):
    x = str(x).zfill(4)  # Ensure the number has 4 digits (e.g., 915 -> 0915)
    hours = int(x[:2])    # Extract hours
    minutes = int(x[2:])  # Extract minutes
    return f"{hours:02d}:{minutes:02d}"

# Apply the conversion function to the ACTTIME column
Times = Times.apply(convert_to_time)

# Drop any rows with invalid or NaN values (if any)
Times = Times.dropna()

# Function to round time to nearest 30 minutes
def round_to_nearest_half_hour(time):
    hours, minutes = map(int, time.split(':'))
    
    # Handle the special case for 24:xx times (next day's 00:xx)
    if hours == 24:
        hours = 0
    
    if minutes < 30:
        # Round down to the nearest half hour
        rounded_time = pd.to_datetime(f"{hours:02d}:00", format='%H:%M').time()
    else:
        # Round up to the next half hour
        rounded_time = pd.to_datetime(f"{hours:02d}:30", format='%H:%M').time()
    
    return str(rounded_time)

# Apply rounding function to all times
rounded_times = Times.apply(round_to_nearest_half_hour)

# Count the occurrences of each rounded time
time_counts = rounded_times.value_counts().sort_index()

# Print the counts of rounded times
print("Rounded Time Counts:\n", time_counts)

# Plotting the number of calls vs time
plt.figure(figsize=(10, 6))
time_counts.plot(kind='bar')
plt.xlabel('Time')
plt.ylabel('Number of Calls')
plt.title('Number of Calls by Rounded Time')
plt.xticks(rotation=45)
plt.show()
```
